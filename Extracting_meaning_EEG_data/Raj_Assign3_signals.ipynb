```python
# Transpose dataframe so that channels are in rows and time samples are in columns
def to_mne_support(df):
  """
  From dataframe to raw mne supported
    
  Parameters:
        df: The dataframe

  returns:
        raw: EEG data with shape (n_channels, n_samples)
  """
  
  df = df.T
  
  # Rename columns with channel names
  channel_names = [f'Channel {i+1}' for i in range(df.shape[1])]
  df.columns = channel_names
  
  # Convert dataframe to MNE RawArray object
  from mne.io import RawArray
  import mne

  # Define sampling frequency (assume 1000 Hz)
  sfreq = 1000
  
  # Define channel types (assume all are EEG)
  ch_types = ['eeg'] * df.shape[1]
  
  # Create MNE info object
  info = mne.create_info(channel_names, sfreq, ch_types)
  
  # Create MNE RawArray object from pandas dataframe
  raw = RawArray(data=df.values.T, info=info)
  montage = mne.channels.make_standard_montage('biosemi32')
  raw.set_montage(montage, on_missing='ignore')
  return raw
```

## Remove redundant

```python
import numpy as np
import mne
def remove_redundant(raw):
  """
  Removes redundant channels
    
  Parameters:
        raw: EEG data with shape (n_channels, n_samples)

  returns:
        raw: EEG data with shape (n_channels, n_samples) without bad or redundant channels
  """
  
  # Remove redundant channels
  montage = mne.channels.make_standard_montage('biosemi32')
  raw.set_montage(montage, on_missing='ignore')
  picks = mne.pick_types(raw.info, eeg=True, exclude='bads')
  raw = raw.pick_channels([raw.ch_names[i] for i in picks])
  return raw
```

## Key channels

### Using the remove redundant channels function from the preceding step, the method first eliminates redundant channels from the raw object. The remaining channels' covariance matrices are then calculated, their diagonal variance extracted, and the channels are sorted by decreasing variance. The key channel(s) are chosen as those with the highest variance, and their names are returned as a list.

```python
def select_key_channels(raw) -> list:
    """
    Identifies the key channels

    Parameters:
        raw: EEG data with shape (n_channels, n_samples)
    
    Returns:
        key_freq_ranges (list): List of channel names that correspond to the frequency range of interest.
    """
    
    # Compute the covariance matrix of the remaining channels
    cov = mne.compute_raw_covariance(raw)

    # Extract the diagonal of the covariance matrix (variance of each channel)
    cov_diag = np.diag(cov.data)

    # Sort the channels by decreasing variance
    sorted_channels = np.argsort(cov_diag)[::-1]

    # Select the channel(s) with the highest activity
    key_channels = [raw.ch_names[sorted_channels[0]]]
    if len(sorted_channels) > 1:
        key_channels.append(raw.ch_names[sorted_channels[1]])

    return key_channels
```

## Key frequencies

### The data's power spectral density (PSD) can be used to determine the important frequency range. Here is a function that chooses one or two channels with the highest activity and returns the frequency range with the highest power that corresponds to those channels.

```python
def select_key_frequencies(raw, key_channels) -> list[tuple]:
    """
    Identifies the key frequencie ranges

    Parameters:
        raw: EEG data with shape (n_channels, n_samples).
        key_channles (list): List of key channels

    Returns:
        key_freq_ranges (list): List of channel names that correspond to the frequency range of interest.
    """

    # Compute the power spectral density (PSD) of the data
    sfreq = raw.info['sfreq']

    # Converting to numpy
    data = raw.get_data()
    psd, freqs = mne.time_frequency.psd_array_welch(data, sfreq=sfreq, fmin=0.5, fmax=40, n_jobs=1)

    # Find the indices of the key channels
    key_indices = [raw.ch_names.index(ch) for ch in key_channels]

    # Find the frequency range with the highest power for each key channel
    key_freq_ranges = []
    for index in key_indices:
        psd_channel = psd[index]
        max_index = np.argmax(psd_channel)
        max_freq = freqs[max_index]
        key_freq_ranges.append((max_freq - 2, max_freq + 2))  # return a range of +/- 2 Hz around the peak frequency

    return key_freq_ranges
```

## Identify cortecies or  cortex channel or channels correspondent to

### In EEG data, the identification of cortical regions or channels that correspond to specific frequencies can be determined through a process called spectral analysis. Spectral analysis involves examining the frequency content of the EEG signal using a mathematical technique called the Fourier transform.

### The function uses the MNE Python package to create a Raw object from the input data and compute the power spectral density (PSD) using the Welch method. The PSD is averaged across time and sorted to identify the channels with the maximum power in the frequency range of interest.

### Finally, the function plots the power spectrum and highlights the frequency range of interest, and returns a list of channel names corresponding to the frequency range.

```python
import matplotlib.pyplot as plt
import mne

def identify_channels_by_frequency(raw, freq_range) -> list:
    """
    Identifies cortical channels that correspond to a specific frequency range in an EEG signal.

    Parameters:
        raw: EEG data with shape (n_channels, n_samples).
        freq_range (tuple): Frequency range of interest as a tuple (low_freq, high_freq).

    Returns:
        channel_names (list): List of channel names that correspond to the frequency range of interest.
    """
    
    # Getting sample frequenices from the raw data info
    sfreq = raw.info['sfreq']

    # Compute power spectral density
    data = raw.get_data()
    fmin, fmax = freq_range
    psd, freqs = mne.time_frequency.psd_array_welch(data, sfreq=sfreq, fmin=fmin, fmax=fmax, n_jobs=1)
    
    psd = np.ravel(psd)
    # Identify channels with maximum power in the frequency range
    power_in_range = np.mean(psd, axis=-1)
    sorted_channels = np.argsort(power_in_range)
    channel_names = [raw.ch_names[i] for i in sorted_channels[::-1]]

    # Plot power spectrum and highlight frequency range
    plt.plot(freqs, power_in_range)
    plt.axvspan(fmin, fmax, alpha=0.5, color='gray')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Power (uV^2/Hz)')
    plt.show()

    return channel_names
```

### The PSD and FFT are tools for measuring and analyzing a signal's frequency content. The FFT transfers time data to the frequency domain, which allows engineers to view changes in frequency values. The PSD takes another step and calculates the power, or strength, of the frequency content.

## Extract slice of key activity

```python
# Select Channel 1 data
def extract_slice(raw_data, freq_range) -> np.ndarray:
  """
    Extract the slices of key frequencies from the channel.

    Parameters:
        raw_data: EEG data with shape (n_channels, n_samples).
        freq_range (tuple): Frequency range of interest as a tuple (low_freq, high_freq).

    Returns:
        filtered_data: The extracted slice
    """

  channel_name = 'Channel 1'
  channel_data = raw_data.copy().pick_channels([channel_name])
  
  # Define the filter parameters
  lowcut, highcut = freq_range
  
  # Apply the bandpass filter to the data
  filtered_data = mne.filter.filter_data(channel_data.get_data()[0], raw_data.info['sfreq'], lowcut, highcut)
  
  # Plot the filtered data
  import matplotlib.pyplot as plt
  plt.plot(filtered_data)
  plt.show()

  return filtered_data
```

we define the filter parameters including the lower and upper cutoff frequencies. Finally, we apply the bandpass filter to the Channel 1 data using the filter_fata() function from MNE's filtering module, and plot the filtered data using the matplotlib library.

# Statistical test

### T-test

### If the p-value is less than the significance level (e.g., 0.05), you can reject the null hypothesis that the means of the two slices are equal and conclude that there is a significant difference between them

```python
from scipy.stats import ttest_ind

def t_test(slice1, slice2) -> None:
  """
  T-test for the slices
  
  Parameters:
      slice1 (np.ndarray) : The extracted slice from the first data
      slice2 (np.ndarray) : The extracted slice from the second data
  """

  # Assume slice1 and slice2 are two arrays of EEG data of key activity
  t_stat, p_value = ttest_ind(slice1, slice2)
  print("t-statistic:", t_stat)
  print("p-value:", p_value)
  
  # If p values < 0.05
  if p_value < 0.05 :
    print("You can reject the null hypothesis that the means of the two slices are equal and conclude that there is a significant difference between them.")

  # If not
  else:
    print("Cannot reject the null hypothesis that the means of the two slices are equal")
```

```python
t_test(first_filtered, second_filtered)
```

### Wilcoxon rank-sum test

### If the p-value is less than the significance level (e.g., 0.05), you can reject the null hypothesis that the distributions of the two slices are equal and conclude that there is a significant difference between them.

```python
from scipy.stats import ranksums

def rank_sum_test(slice1, slice2) -> None:
  """
  T-test for the slices
  
  Parameters:
      slice1 (np.ndarray) : The extracted slice from the first data
      slice2 (np.ndarray) : The extracted slice from the second data
  """
  
  # Assume slice1 and slice2 are two arrays of EEG data of key activity
  U_stat, p_value = ranksums(slice1, slice2)
  print("U-statistic:", U_stat)
  print("p-value:", p_value)
 
  # If p-value is less than 0.05
  if p_value < 0.05:
     print("Reject the null hypothesis that the means of the two slices are equal and conclude that there is a significant difference between them.")

  # If it is not
  else:
    print("Cannot reject null hypothesis, the distributions of the two slices are equal")
 
```

```python
rank_sum_test(first_filtered, second_filtered)
```
